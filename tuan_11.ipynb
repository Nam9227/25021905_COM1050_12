```python
def valid_regex(s):
    stack = []
    prev = ""

    for i, c in enumerate(s):
        if c == '(':
            if prev.isalpha() or prev == ')' or prev == '*':
                return False
            stack.append(c)

        elif c == ')':
            if not stack or prev in '|(':
                return False
            stack.pop()

        elif c == '|':
            if i == 0 or i == len(s)-1 or prev in '|(':
                return False

        elif c == '*':
            if i == 0 or prev in '|(':
                return False

        elif c.isalpha():
            if prev == ')':
                return False

        else:
            return False  # ký tự lạ

        prev = c

    return len(stack) == 0 and prev not in '|('
```

```python
t = int(input().strip())
for _ in range(t):
    s = input().strip()
    valid = True
    prev_repeat = False   
    i = 0
    while i < len(s):
        if s[i] == '\\':         
            if i == len(s) - 1:  
                valid = False
                break
            prev_repeat = False
            i += 2
            continue
        if s[i] in '*+':         
            if prev_repeat:      
                valid = False
                break
            prev_repeat = True
        else:
            prev_repeat = False
        i += 1
    print("True" if valid else "False")    
```
