```python
def binary_search(a,d,c,k):
    while d<=c:
        g=int((d+c)//2)
        if k>a[g]:
            d=g+1
        elif k<a[g]:
            c=g-1
        elif k==a[g]:
            return g+1
    return -1
a=list(map(int,input().split()))
k=int(input())
print(binary_search(a,0,len(a)-1,k))

```

```python
def sx_noibot(a):
    for i in range(len(a)):
        b=True
        for j in range(1,len(a)-i):
            if a[j]<a[j-1]:
                a[j],a[j-1]=a[j-1],a[j]
                b=False
        if b :
            break
    return a
a=list(map(int,input().split()))
a=sx_noibot(a)
for i in a:
    print(i,end=' ')
```

```python
def sx_chon(a):
    for i in range(len(a)-1):
        k=i
        th=a[i]
        for j in range(i+1,len(a)):
            if a[j]<=th:
                th=a[j]
                k=j 
        a[i],a[k]=a[k],a[i]
    return a
a=list(map(int,input().split()))
a=sx_chon(a)
for i in a:
    print(i,end=' ')
```
