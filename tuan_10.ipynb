```python
a=list(map(int,input().split()))
b=list(map(int,input().split()))
i,j=0,0
c=[]
a1,b1=0,0
while a1==0 or b1==0 :
    k=0

    if (a[i]<=a[j] and a1==0) or b1==1:
        if i==len(a)-1:
            a1=1
        c.append(a[i])
        k=1
    else: 
        if j==len(b)-1:
            b1=1
        c.append(b[j])
    if k==0 :
        j+=1 if a1==0 else 0
    else:
        i+=1 if a1==0 else 0
print(*c)
```
