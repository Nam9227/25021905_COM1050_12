```python
def Median(nums):
    a, b, c, d, e = nums

    # So sánh cặp (a, b)
    if a > b:
        x1, y1 = a, b
    else:
        x1, y1 = b, a

    # So sánh cặp (c, d)
    if c > d:
        x2, y2 = c, d
    else:
        x2, y2 = d, c

    # So sánh hai số lớn hơn
    if x1 > x2:
        L = x1
        M1 = x2
    else:
        L = x2
        M1 = x1

    # So sánh hai số nhỏ hơn
    if y1 > y2:
        M2 = y1
    else:
        M2 = y2

    # Trung vị của 3 số M1, M2, e
    if e > M1:
        if e < M2:
            return e
        else:
            return M2
    else:
        if M1 < M2:
            return M1
        else:
            return M2
a=list(map(int,input().split()))
Median(a)
```
