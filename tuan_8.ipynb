```python
class ShoppingCart:
    def __init__(self):
        self.sp = []

    def add_item(self, ten, gia):
        self.sp.append({"ten": ten, "gia": gia})
        print("Đã thêm sản phẩm.")

    def remove_item(self, ten):
        for i in self.sp:
            if i["ten"] == ten:
                self.items.remove(i)
                print("Đã xóa sản phẩm.")
                return
        print("Không tìm thấy sản phẩm.")

    def is_empty(self):
        return len(self.sp) == 0

    def total_price(self):
        k = 0
        for i in self.sp:
            k += i["gia"]
        return k

    def show_items(self):
        if self.is_empty():
            print("Giỏ hàng rỗng.")
        else:
            print("Danh sách sản phẩm:")
            for i in self.sp:
                print(f"- {i['ten']} : {i['gia']}")

    def clear_cart(self):
        self.sp.clear()
        print("Đã xóa toàn bộ giỏ hàng.")
```

```python
cart = ShoppingCart()

while True:
    print("\n--- MENU ---")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = input("Nhập lựa chọn (2-7): ")

    if choice == "2":
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        cart.add_item(name, price)

    elif choice == "3":
        name = input("Nhập tên sản phẩm cần xóa: ")
        cart.remove_item(name)

    elif choice == "4":
        if cart.is_empty():
            print("Giỏ hàng đang rỗng.")
        else:
            print("Giỏ hàng không rỗng.")

    elif choice == "5":
        print("Tổng tiền:", cart.total_price())

    elif choice == "6":
        cart.show_items()

    elif choice == "7":
        cart.clear_cart()

    else:
        print("Lựa chọn không hợp lệ.")

    exit_choice = input("Có muốn thoát khỏi chương trình không? (yes/no): ")
    if exit_choice.lower() == "yes":
        print("Đã thoát chương trình.")
        break
```

```python
import math
class diem():
    def __init__(self,a=0.0,b=0.0,c=0.0):
        self.x=a
        self.y=b
        self.z=c
    def __str__(self):
        return f" {self.x}   {self.y}   {self.z} "
def tao(a):
    vt1=diem(a[1].x-a[0].x,a[1].y-a[0].y,a[1].z-a[0].z)
    vt2=diem(a[2].x-a[0].x,a[2].y-a[0].y,a[2].z-a[0].z)
    #print(vt1,vt2)
    n1=diem(vt1.y * vt2.z - vt1.z * vt2.y , vt1.z * vt2.x - vt1.x * vt2.z , vt1.x * vt2.y - vt1.y * vt2.x)
    #print(n1)
    return n1
def nhan(vt1,vt2):
    #print(vt1,vt2)
    a=abs(vt1.x*vt2.x+vt1.y*vt2.y+vt1.z*vt2.z)
    b=(vt1.x**2+vt1.y**2+vt1.z**2)**0.5*(vt2.x**2+vt2.y**2+vt2.z**2)**0.5
    #print(a,b)
    return a/b if b!=0 else 1
m1=[]
m2=[]
for i in range(4):
    a=list(map(float,input().split()))
    c=diem(a[0],a[1],a[2])
    if 0<=i<=2:
        m1.append(c)
    if 1<=i<=3:
        m2.append(c)
#print(*m1)
#print(*m2)
n1=tao(m1)
n2=tao(m2)
g=nhan(n1,n2)
#print(g)
g=abs(g)
g=math.degrees(math.acos(g))
print(f"{g:.2f}")
```
