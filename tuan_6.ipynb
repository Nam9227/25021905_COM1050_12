```python
kq = []
c = {}
a = list(map(int, input().split()))
k = int(input())
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] == k:
            pair = (a[i], a[j])
            if pair not in kq:
                kq.append(pair)
kq.sort(key=lambda t: t[0])
print(kq)
```

```python
from itertools import combinations

a = list(map(int, input().split()))
k = int(input())

result = sorted(
    set((x, y) for x, y in combinations(a, 2) if x + y == k),
    key=lambda t: t[0]
)

print(result)
```
