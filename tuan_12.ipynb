```python
def reverse(a: str) -> int:
    try:
        with open(a,'r') as f:
            a=f.read()
        return int(a[::-1] if a[0] != '-' else a[0]+a[:0:-1])
    except:
        with open(a,'w') as f:
            f.write("")
        a="File Ko Tồn Tại"
        return a
a=input()
print(reverse(a))
```

```python
import unicodedata

def xuly_chuoi(s):
    s = s.lower()
    s = unicodedata.normalize('NFD', s)
    s = ''.join(c for c in s if unicodedata.category(c) != 'Mn')
    s = s.replace('đ', 'd')
    return s

def lm(a,b):
    s,s1="",""
    try:
        with open(a,'r') as f:
            s=f.read()
        with open(b,'r') as f:
            s1=f.read()   
    except:
        return "loi doc file"
    s=xuly_chuoi(s)
    s1=xuly_chuoi(s1)
    b,b1={},{}
    for i in s:
        if i!=' ':
            b[i] = b.get(i,0) + 1
    for i in s1:
        if i!=' ':
            b1[i] = b1.get(i,0) + 1
    return b1==b
a=input()
b=input()
print(lm(a,b))
```

```python
def lm(a):
    try:
        with open(a, 'r') as f:
            b = [i.strip() for i in f.readlines() if i.strip()]
            
        for i in range(0, len(b), 2):
            loai_hinh = b[i].upper()
            thong_so = b[i+1].upper()
            chu_vi = 0.0
            if "SQUARE" in loai_hinh:
                side = float(thong_so.split("SIDE:")[1].strip())
                chu_vi = 4 * side
            elif "RECTANGLE" in loai_hinh:
                parts = thong_so.split()
                w = float(parts[1]) 
                h = float(parts[3]) 
                chu_vi = 2 * (w + h)
            elif "CIRCLE" in loai_hinh:
                r = float(thong_so.split("RADIUS:")[1].strip())
                chu_vi = 2 *3.1415* r
            print(f"{chu_vi:.2f}")
    except FileNotFoundError:
        print("Mission failed: Không tìm thấy tệp.")
    except Exception as e:
        print(f"Mission failed: {e}")

a=input()
lm(a)
```

```python
def lm(a):
    try:
        with open(a, 'r') as f:
            c = [i.strip().upper() for i in f.readlines() if i.strip()]
            x1, y1 = 0.0, 0.0
            x2, y2 = 0.0, 0.0
            v = 0.0
            
            for i in c:
                b = i.split()
                if "SRC" in i:
                    x1 = float(b[2])
                    y1 = float(b[4])
                elif "DES" in i:
                    x2 = float(b[2])
                    y2 = float(b[4])
                elif "VELOCITY" in i:
                    v = float(b[1])
            
            if v == 0:
                return 0.0
            s=((x2-x1)**2+(y2-y1)**2)**0.5
            t=s/v
            return round(t,2)
            
    except Exception:
        print("Mission failed")
        return 0.0
a=input()
print(lm(a))
```

```python
def lm(path_file):
    try:
        with open(path_file, 'r') as f:
            data = [line.split() for line in f if line.strip()]
        if not data:
            return
        n = len(data)
        m = [[int(x) for x in row] for row in data]
        kq = [[0] * n for _ in range(n)]
        for r in range(n):
            for c in range(n):
                tong = 0
                so_luong = 0
                for i in range(r - 1, r + 2):
                    for j in range(c - 1, c + 2):
                        if 0 <= i < n and 0 <= j < n:
                            tong += m[i][j]
                            so_luong += 1
                kq[r][c]=tong //so_luong
        for i in kq:
            print(*i)
    except Exception:
        print("Mission failed")

a = input()
lm(a)
```
