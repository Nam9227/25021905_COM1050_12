```python
def reverse(a: str) -> int:
    try:
        with open(a,'r') as f:
            a=f.read()
        return int(a[::-1] if a[0] != '-' else a[0]+a[:0:-1])
    except:
        with open(a,'w') as f:
            f.write("")
        a="File Ko Tồn Tại"
        return a
a=input()
print(reverse(a))
```

```python
def canWinNim (a: str):
    try:
        with open(a,'r') as f:
            a=f.read()
        a=int(a)
        if a%4==0:
            return False
        else:
            return True
    except:
        with open(a,'w') as f:
            f.write("")
        return False
a=input()
if canWinNim(a):
    print("win r do")
else:
    print("het cuu")
```

```python
import unicodedata

def xuly_chuoi(s):
    s = s.lower()
    s = unicodedata.normalize('NFD', s)
    s = ''.join(c for c in s if unicodedata.category(c) != 'Mn')
    s = s.replace('đ', 'd')
    return s

def lm(a,b):
    s,s1="",""
    try:
        with open(a,'r') as f:
            s=f.read()
        with open(b,'r') as f:
            s1=f.read()   
    except:
        return "loi doc file"
    s=xuly_chuoi(s)
    s1=xuly_chuoi(s1)
    b,b1={},{}
    for i in s:
        if i!=' ':
            b[i] = b.get(i,0) + 1
    for i in s1:
        if i!=' ':
            b1[i] = b1.get(i,0) + 1
    return b1==b
a=input()
b=input()
print(lm(a,b))
```

```python
def lm(n):
    f=[""]*(n+1)
    f[0]="a"
    f[1]="b"
    for i in range(2,n):
        f[i]=f[i-1]+f[i-2]
    with open("output.txt",'w') as f1:
        for i in range(n):
            f1.write(f"f({i+1}) = {f[i]}\n")
a=int(input())
lm(a)
```

```python
def lm(a):
    b=[]
    try:
        with open(a,'r') as f:
            b=list(map(int,f.read().split()))
    except:
        print("loi doc file")
    d=[0]*(len(b))
    for i in range(len(b)):
        d[i]=b[i]
        for j in range(0,i-1):
            d[i]=max(d[i],d[j]+b[i])
    print(max(d))
a=input()
lm(a)
```
